```python
conn = sqlite3.connect("hr.db")

pd.read_sql(
    "SELECT COUNT(*) AS total_employees FROM employees",
    conn
)
```

```python
df["department"].value_counts()
```

```python
pd.read_sql("""
SELECT department, COUNT(*) AS employee_count
FROM employees
GROUP BY department
ORDER BY employee_count DESC
""", conn)
```

```python
df.groupby("jobrole")["monthlyincome"].mean().sort_values(ascending=False)
```

```python
pd.read_sql("""
SELECT jobrole, ROUND(AVG(monthlyincome), 2) AS avg_monthly_income
FROM employees
GROUP BY jobrole
ORDER BY avg_monthly_income DESC
""", conn)
```

```python
df.sort_values(
    ["performancerating", "monthlyincome"],
    ascending=[False, False]
).head(5)[
    ["employeenumber", "jobrole", "department", "performancerating", "monthlyincome"]
]
```

```python
pd.read_sql("""
SELECT employeenumber,
       jobrole,
       department,
       performancerating,
       monthlyincome
FROM employees
ORDER BY performancerating DESC, monthlyincome DESC
LIMIT 5
""", conn)
```

```python
df.groupby("department")["performancerating"].mean().sort_values(ascending=False)
```

```python
pd.read_sql("""
SELECT department,
       ROUND(AVG(performancerating), 3) AS avg_performance_rating
FROM employees
GROUP BY department
ORDER BY avg_performance_rating DESC
LIMIT 1
""", conn)
```

```python
df.groupby("department")["attrition"] \
  .apply(lambda x: (x == "Yes").mean()) \
  .sort_values(ascending=False)
```

```python
pd.read_sql("""
SELECT department,
       AVG(CASE WHEN attrition = 'Yes' THEN 1 ELSE 0 END) AS attrition_rate
FROM employees
GROUP BY department
ORDER BY attrition_rate DESC
""", conn)
```

```python
df.groupby("overtime")["attrition"] \
  .apply(lambda x: (x == "Yes").mean())
```

```python
pd.read_sql("""
SELECT overtime,
       AVG(CASE WHEN attrition = 'Yes' THEN 1 ELSE 0 END) AS attrition_rate
FROM employees
GROUP BY overtime
""", conn)
```

```python
df.groupby("education")["monthlyincome"] \
  .mean() \
  .sort_values(ascending=False)
```

```python
pd.read_sql("""
SELECT education,
       ROUND(AVG(monthlyincome), 2) AS avg_monthly_income
FROM employees
GROUP BY education
ORDER BY avg_monthly_income DESC
""", conn)
```
